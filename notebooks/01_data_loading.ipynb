# 01. 데이터 로딩

원천 데이터를 읽고 전처리하여 시군구 단위 기본 데이터를 구성합니다.

```python
import sys
sys.path.insert(0, '../src')
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 120)
print('환경 로드 완료')
```

## 1-1. 원천 파일 목록

```python
from pathlib import Path
raw_dir = Path('../data/raw')
for f in sorted(raw_dir.iterdir()):
    if f.is_file():
        size_kb = f.stat().st_size / 1024
        print(f'{f.name:50s}  {size_kb:8.0f} KB')
```

## 1-2. 문화행사 데이터

```python
from preprocessing import standardize_culture_events
events = standardize_culture_events()
print(f'행 수: {len(events):,}  / 컬럼: {list(events.columns)}')
events.head(5)
```

```python
print('시도별 행사 수:')
events.groupby('province')['event_name'].count().sort_values(ascending=False)
```

## 1-3. 박물관·미술관 데이터

```python
from preprocessing import standardize_museums
museums = standardize_museums()
print(f'행 수: {len(museums):,}')
print('시설 유형:', museums['facility_type'].value_counts().head(5).to_dict())
museums.head(5)
```

## 1-4. 지역축제 데이터

```python
from preprocessing import standardize_festivals
festivals = standardize_festivals()
print(f'행 수: {len(festivals):,}  / 시군구: {festivals["sigungu"].nunique()}개')
print('상위 시군구:\n', festivals.groupby(["province","sigungu"]).size().sort_values(ascending=False).head(10))
```

## 1-5. 버스정류장 데이터

```python
from preprocessing import standardize_bus_stops
bus = standardize_bus_stops()
print(f'행 수: {len(bus):,}  / 고유 지역: {bus["region_name"].nunique()}개')
bus.groupby('province')['bus_stop_id'].count().sort_values(ascending=False)
```

## 1-6. 시군구 단위 인구 (시도 비례 배분)

```python
from preprocessing import standardize_population
pop = standardize_population()
print(f'시군구 수: {len(pop)}개')
print('점수 상위 인구 지역:')
pop.sort_values('population', ascending=False).head(10)[['region_name','population','youth_ratio','elderly_ratio']]
```
