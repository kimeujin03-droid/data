# 02. 피처 엔지니어링

시군구별로 문화공급·수요·접근성 변수를 계산합니다.

```python
import sys
sys.path.insert(0, '../src')
import pandas as pd, warnings
warnings.filterwarnings('ignore')
from preprocessing import load_raw_data
from feature_engineering import build_region_features_from_raw
raw = load_raw_data()
features = build_region_features_from_raw(raw)
print(f'시군구: {len(features)}개 / 피처: {len(features.columns)}개')
```

## 2-1. 핵심 변수 분포

```python
key_cols = ['facility_per_10k','event_per_month','museum_count',
            'festival_count','bus_stop_count','elderly_ratio','youth_ratio']
features[key_cols].describe().round(2)
```

## 2-2. 문화공급 최상위 시군구

```python
features.sort_values('event_per_month', ascending=False).head(10)
[['region_name','population','event_per_month','museum_count','festival_count','bus_stop_count']]
```

## 2-3. 접근성 최취약 시군구 (교통 취약 상위)

```python
features.sort_values('transport_weakness', ascending=False).head(10)
[['region_name','province','transport_weakness','rural_flag','elderly_ratio']]
```

## 2-4. 프로그램 미스매치

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
features['elderly_program_ratio'].hist(bins=20, ax=axes[0], color='#e8956b')
axes[0].set_title('고령층 친화 프로그램 비율')
axes[0].set_xlabel('비율')
features['youth_program_ratio'].hist(bins=20, ax=axes[1], color='#5b8fde')
axes[1].set_title('청소년 교육형 프로그램 비율')
axes[1].set_xlabel('비율')
plt.tight_layout()
plt.show()
```

## 2-5. 피처 상관관계

```python
corr_cols = ['facility_per_10k','event_per_month','elderly_ratio',
             'youth_ratio','transport_weakness','mismatch_raw']
features[corr_cols].corr().round(2)
```
